**Lab Task 1: Basic Feistel Cipher Implementation**

```python
# Basic Feistel Cipher - Encryption Functions
def xor(a, b):
    """Perform XOR operation on two binary strings"""
    return ''.join(str(int(x) ^ int(y)) for x, y in zip(a, b))

def feistel_round(left, right, key):
    """Single Feistel round"""
    f_output = xor(right, key)
    new_left = right
    new_right = xor(left, f_output)
    return new_left, new_right, f_output

def encrypt_feistel(plaintext, keys):
    """Encrypt using 4-round Feistel cipher"""
    # Split plaintext into left and right halves
    left, right = plaintext[:4], plaintext[4:]

    print(" ENCRYPTION PROCESS ")
    print(f"Initial: L0={left}, R0={right}")
    print("\n")

    rounds = []
    for i, key in enumerate(keys, 1):
        new_left, new_right, f_output = feistel_round(left, right, key)
        rounds.append((i, left, right, f_output, new_left, new_right))

        print(f"Round {i}:")
        print(f"  L{i-1}={left}, R{i-1}={right}")
        print(f"  F(R{i-1},K{i})={f_output}")
        print(f"  L{i}={new_left}, R{i}={new_right}")
        print()

        left, right = new_left, new_right

    ciphertext = left + right
    return ciphertext, rounds
```

```python
# Basic Feistel Cipher - Decryption Function
def decrypt_feistel(ciphertext, keys):
    """Decrypt using reverse round keys"""
    # Split ciphertext into left and right halves
    left, right = ciphertext[:4], ciphertext[4:]

    print(" DECRYPTION PROCESS ")
    print(f"Initial: L4={left}, R4={right}")
    print("\n")

    # Reverse keys for decryption
    rev_keys = keys[::-1]

    for i, key in enumerate(rev_keys, 1):
        f_output = xor(right, key)
        new_left = xor(f_output, left)
        new_right = left

        round_num = 5 - i
        print(f"Round {i} (Reverse K{5-i}):")
        print(f"  L4={left}, R4={right}")
        print(f"  F(R4,K{5-i})={f_output}")
        print(f"  L{round_num-1}={new_left}, R{round_num-1}={new_right}")
        print()

        left, right = new_left, new_right

    plaintext = left + right
    return plaintext
```

```python
# Basic Feistel Cipher - Main Execution
# Sample input
plaintext = "11001100"
keys = ["1010", "0101", "1111", "0001"]

print("Lab Task 1: Basic Feistel Cipher")
print(f"Plaintext: {plaintext}")
print(f"Keys: K1={keys[0]}, K2={keys[1]}, K3={keys[2]}, K4={keys[3]}")
print("\n")

# Encrypt
ciphertext, enc_rounds = encrypt_feistel(plaintext, keys)
print(f"Ciphertext: {ciphertext}")
print("\n")

# Decrypt
decrypted = decrypt_feistel(ciphertext, keys)
print(f"Decrypted: {decrypted}")
print(f"Verification: {'LESSSGOOO PASS' if decrypted == plaintext else '✗ FAIL'}")
```

Lab Task 2

Feistel Cipher with Custom Round Function

```python
# Task 2  Custom Round Function (Mod 16 Addition)
def mod16_add_round(right, key):
    """Round function: (R + K) mod 16"""
    r_val = int(right, 2)
    k_val = int(key, 2)
    result = (r_val + k_val) % 16
    return format(result, '04b')

def feistel_mod16_round(left, right, key):
    """Feistel round with mod16 addition"""
    f_output = mod16_add_round(right, key)
    new_left = right
    new_right = xor(left, f_output)
    return new_left, new_right, f_output

def encrypt_mod16(plaintext, keys):
    """Encrypt with mod16 round function"""
    left, right = plaintext[:4], plaintext[4:]

    print(" MOD16 FEISTEL ENCRYPTION ")
    print(f"Initial: L0={left}, R0={right}")
    print("\n")

    for i, key in enumerate(keys, 1):
        new_left, new_right, f_output = feistel_mod16_round(left, right, key)
        print(f"Round {i} (K{i}={key}):")
        print(f"  L{i-1}={left}, R{i-1}={right}")
        print(f"  F={f_output} = ({int(right,2)}+{int(key,2)}) mod 16")
        print(f"  L{i}={new_left}, R{i}={new_right}")
        print()
        left, right = new_left, new_right

    return left + right
```

```python
# Task 2  Complete Execution
plaintext = "11001100"
keys = ["1010", "0101", "1111", "0001"]

print("Lab Task 2: Feistel with Mod16 Round Function")
print(f"Plaintext: {plaintext}")
print("Keys:", keys)
print("\n")

ciphertext = encrypt_mod16(plaintext, keys)
print(f"Ciphertext: {ciphertext}")

# Decryption (reuses Task 1 decrypt with mod16 rounds)
decrypted = decrypt_feistel(ciphertext, keys)  # Works due to Feistel structure
print(f"Decrypted: {decrypted}")
print(f"Verification: {'✓ PASS' if decrypted == plaintext else '✗ FAIL'}")

print("\nWhy it works: Feistel structure is reversible when keys are applied in reverse order!")
```

Lab Task 3

Feistel Structure Using ASCII Characters

```python
import random

def text_to_bin(text):
    """Convert text to 8-bit binary representation"""
    result = []
    for char in text:
        # Get ASCII value
        ascii_val = ord(char)
        # Convert to 8-bit binary
        binary = ''
        for i in range(7, -1, -1):
            if ascii_val & (1 << i):
                binary += '1'
            else:
                binary += '0'
        result.append(binary)
    return ''.join(result)

def bin_to_text(binary_str):
    """Convert binary string back to text"""
    # Make sure length is multiple of 8
    if len(binary_str) % 8 != 0:
        return ""

    result = []
    for i in range(0, len(binary_str), 8):
        byte = binary_str[i:i+8]
        # Convert binary to ASCII
        value = 0
        for bit in byte:
            value = (value << 1) | (1 if bit == '1' else 0)
        result.append(chr(value))
    return ''.join(result)

def xor_4bit(a, b):
    """XOR two 4-bit binary strings"""
    result = ''
    for i in range(4):
        result += '1' if a[i] != b[i] else '0'
    return result

def feistel_function(right_half, key):
    """Simple Feistel function - XOR with key after converting to int"""
    # Convert 4-bit binary to int
    right_int = 0
    for bit in right_half:
        right_int = (right_int << 1) | (1 if bit == '1' else 0)

    # XOR with key
    result_int = right_int ^ key

    # Convert back to 4-bit binary
    result = ''
    for i in range(3, -1, -1):
        if result_int & (1 << i):
            result += '1'
        else:
            result += '0'
    return result

def encrypt_ascii(plaintext):
    """Encrypt ASCII text using 4-round Feistel cipher"""
    # Convert text to binary
    binary_text = text_to_bin(plaintext)

    # Split into 8-bit blocks
    blocks = []
    for i in range(0, len(binary_text), 8):
        if i + 8 <= len(binary_text):
            blocks.append(binary_text[i:i+8])

    cipher_blocks = []
    keys = []

    for block in blocks:
        # Split into left and right halves (4 bits each)
        L = block[:4]
        R = block[4:]

        # Generate random key for this block (0-15)
        key = random.randint(0, 15)
        keys.append(key)

        # 4 rounds of Feistel network
        for round_num in range(4):
            # Apply Feistel function to right half
            f_result = feistel_function(R, key)

            # New right = left XOR f_result
            new_R = xor_4bit(L, f_result)

            # New left = old right
            new_L = R

            # Update for next round
            L, R = new_L, new_R

        # Combine final halves
        cipher_block = L + R
        cipher_blocks.append(cipher_block)

    # Return concatenated binary string and keys
    return ''.join(cipher_blocks), keys

def decrypt_ascii(cipher_binary, keys):
    """Decrypt Feistel cipher text"""
    blocks = []
    for i in range(0, len(cipher_binary), 8):
        if i + 8 <= len(cipher_binary):
            blocks.append(cipher_binary[i:i+8])

    plain_blocks = []

    for i, block in enumerate(blocks):
        if i >= len(keys):
            break

        L = block[:4]
        R = block[4:]
        key = keys[i]

        # Reverse 4 rounds of Feistel
        for round_num in range(4):
            f_result = feistel_function(L, key)
            new_R = xor_4bit(R, f_result)  # recover original L
            new_L = L                       # recover original R
            L, R = new_R, new_L

        # No swap needed — L and R are back to original positions
        plain_blocks.append(L + R)

    return bin_to_text(''.join(plain_blocks))

# Debug function to verify binary conversion
def debug_binary_conversion():
    print("Debug: Binary conversion test")
    test_text = "HI"
    binary = text_to_bin(test_text)
    print(f"'{test_text}' -> binary: {binary} (length: {len(binary)})")
    text_back = bin_to_text(binary)
    print(f"Binary -> text: '{text_back}'")
    print(f"Success: {test_text == text_back}")
    print()

# Run debug first
debug_binary_conversion()

# Test the implementation
print("Lab Task 3: ASCII Feistel Cipher")
print("\n")

# Test 1: "HI"
plaintext = "HI"
print(f"Original text: '{plaintext}'")
print(f"Binary: {text_to_bin(plaintext)}")

ciphertext_bin, keys = encrypt_ascii(plaintext)
print(f"Encrypted binary: {ciphertext_bin}")
print(f"Keys used: {keys}")
print(f"Encrypted length: {len(ciphertext_bin)} bits (should be {len(plaintext)*8})")

decrypted = decrypt_ascii(ciphertext_bin, keys)
print(f"Decrypted text: '{decrypted}'")
print(f"Decryption {'✓ PASSED' if decrypted == plaintext else '✗ FAILED'}")
print()

# Test 2: "Hello"
plaintext2 = "Hello"
print(f"Original text: '{plaintext2}'")
print(f"Binary: {text_to_bin(plaintext2)}")

ciphertext_bin2, keys2 = encrypt_ascii(plaintext2)
print(f"Encrypted binary: {ciphertext_bin2}")
print(f"Keys used: {keys2}")
print(f"Encrypted length: {len(ciphertext_bin2)} bits (should be {len(plaintext2)*8})")

decrypted2 = decrypt_ascii(ciphertext_bin2, keys2)
print(f"Decrypted text: '{decrypted2}'")
print(f"Decryption {'✓ PASSED' if decrypted2 == plaintext2 else '✗ FAILED'}")
print()

# Test 3: Longer text
plaintext3 = "Cryptography"
print(f"Test 3: '{plaintext3}'")
ciphertext_bin3, keys3 = encrypt_ascii(plaintext3)
decrypted3 = decrypt_ascii(ciphertext_bin3, keys3)
print(f"Decrypted: '{decrypted3}'")
print(f"Decryption {'✓ PASSED' if decrypted3 == plaintext3 else '✗ FAILED'}")

# Test 4: Empty string
plaintext4 = ""
print(f"\nTest 4: Empty string")
ciphertext_bin4, keys4 = encrypt_ascii(plaintext4)
decrypted4 = decrypt_ascii(ciphertext_bin4, keys4)
print(f"Decrypted: '{decrypted4}'")
print(f"Decryption {'✓ PASSED' if decrypted4 == plaintext4 else '✗ FAILED'}")
```

Lab Task 4

 Avalanche Effect

```python
# Task 4 - Avalanche Effect Analysis
def full_feistel_encrypt(plaintext, keys):
    """Complete 4-round Feistel encryption"""
    left, right = plaintext[:4], plaintext[4:]
    for key in keys:
        f_output = xor(right, key)
        left, right = right, xor(left, f_output)
    return left + right

def hamming_distance(a, b):
    """Count differing bits between two binary strings"""
    return sum(c1 != c2 for c1, c2 in zip(a, b))

keys = ["1010", "0101", "1111", "0001"]

print("Lab Task 4: AVALANCHE EFFECT")
print("\n")

p1 = "10101010"  # Original
p2 = "10101011"  # 1-bit change

c1 = full_feistel_encrypt(p1, keys)
c2 = full_feistel_encrypt(p2, keys)

print(f"P1: {p1} → C1: {c1}")
print(f"P2: {p2} → C2: {c2}")
print(f"Input diff: {hamming_distance(p1, p2)} bits")
print(f"Output diff: {hamming_distance(c1, c2)} bits")
print(f"Avalanche: {hamming_distance(c1,c2)/8*100:.1f}% bits changed")

print("\nAvalanche Effect: Small input change → Large output change")
print("Importance: Makes cryptanalysis difficult")
print("More rounds → Better diffusion")
```

User-Defined Rounds

```python
# Task 5 Dynamic Feistel Cipher
class DynamicFeistel:
    def __init__(self):
        self.round_outputs = []

    def encrypt(self, plaintext, keys, num_rounds):
        """Dynamic encryption with user-defined rounds"""
        self.round_outputs = []
        left, right = plaintext[:4], plaintext[4:]

        print(f"DYNAMIC FEISTEL ({num_rounds} ROUNDS) ")
        print(f"Plaintext: {plaintext}")
        print("Round | Left | Right | F(R,K) |")
        print("\n")

        for i in range(num_rounds):
            key = keys[i % len(keys)]  # Cycle keys if needed
            f_output = xor(right, key)
            new_left, new_right = right, xor(left, f_output)

            self.round_outputs.append((left, right, f_output))
            print(f"{i+1:5d} | {left:4} | {right:4} | {f_output:4} |")

            left, right = new_left, new_right

        return left + right

    def decrypt(self, ciphertext, keys, num_rounds):
        """Dynamic decryption"""
        left, right = ciphertext[:4], ciphertext[4:]
        rev_keys = keys[::-1]

        for i in range(num_rounds):
            key = rev_keys[i % len(rev_keys)]
            f_output = xor(right, key)
            new_left = xor(f_output, left)
            new_right = left
            left, right = new_left, new_right

        return left + right

# Test with different round counts
print("\n")
feistel = DynamicFeistel()
keys = ["1010", "0101", "1111", "0001"]
plaintext = "11001100"

for rounds in [2, 4, 6]:
    ciphertext = feistel.encrypt(plaintext, keys, rounds)
    decrypted = feistel.decrypt(ciphertext, keys, rounds)
    print(f"\n{rounds} rounds: {plaintext} → {ciphertext} → {decrypted}")
    print(f" LESSGOO {'PASS' if decrypted == plaintext else 'FAIL'}")
```
